### 문제 5번 

다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addall(list):
    sum = 0
    for i in list:
        sum += i
    print(sum)
```

```python
addall([1])
```

```python
addall([1,2,3,4,5,6,7,8,9])
```

addall 함수 내부에는 for문을 이용하여 각각의 원소에 접근하여 모든 값을 sum이라는 변수에 더하여 저장한다.

2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
def addall(list):
    print(sum(list))
```

```python
addall([1,4])
```

addall함수에 sum()함수를 이용하여 list의 모든 원소들을 더해준다.

리스트 내포를 이용하라고 했는데 이번 문제에서는 사용할 필요가 없을 것 같다. 

리스트 변하는 것이 아니기 때문에 내포를 이용하여 하지 않아도 된다.

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addallodd(list):
    sum = 0
    for i in list:
        if i % 2:
            sum += i
    print(sum)
```

```python
addallodd([1])
```

```python
addallodd([1,2,3,4,5,6,7,8,9])
```

```python
def addallodd(list):
    print(sum([ i for i in list if i % 2]))
```

```python
addallodd([1,2,3,4,5,6,7,8,9])
```

그것을 sum()를 이용하여 모든 원소의 합을 출력한다.

```python
k = list(d.keys())
k.sort()
print(k)
for keys in k:
    print((keys, d[keys]))
```

사전 d의 키들 리스트로 만들어서 정렬한다.

그런 다음 정렬된 키의 순서대로 키에 해당하는 값을 불러와 출력한다.

```python
k = list(d.keys())
k.sort(reverse=True)
print(k)
for keys in k:
    print((keys, d[keys]))
```

위와 같은 방법으로 하지만 내림차순으로 정렬해야 하므로 reverse = True 로 설정해 주어야 한다.

일치할 경우 출력하도록 하였다.

```python
k = list(d.values())
k.sort(reverse = True)
for values in k:
    for keys in d:
        if(d[keys] == values):
            print((keys, d[keys]))
            break;
```

```python
from urllib import request
import string
 
url = 'http://www.naver.com'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
#print(source)
print("------------------------")

# 주석태그와 다른 문자가 붙어있지 않도록 나눈다.
source = source.replace("\n"," ")
source = source.replace("\t"," ")
source = source.replace("<!--"," <!-- ")
source = source.replace("-->"," --> ")
source = source.replace("<script"," <script ")
source = source.replace("</script>"," </script> ")
source = source.replace("<style"," <style ")
source = source.replace("</style>"," </style> ")


source = source.split(" ")

source_real = []
for i in source:
    i = i.strip()
    if i != "":
        source_real.append(i)
        
print(source_real)
```

```python
#js,css 태그 제거
first = []
second = []
status = False
for i in source_real:
    
    if "<script" in i:
        if status == False:
            first.append(source_real.index("<script"))
        source_real[source_real.index("<script")] = '@'
        status = True
    if "</script>" in i:
        if status == True:
            second.append(source_real.index("</script>"))
        source_real[source_real.index("</script>")] = '%'
        status = False
        
    if "<style" in i:
        if status == False:
            first.append(source_real.index("<style"))
        source_real[source_real.index("<style")] = '@'
        status = True
    if "</style>" in i:
        if status == True:
            second.append(source_real.index("</style>"))
        source_real[source_real.index("</style>")] = '%'
        status = False


firstlen = len(first)

for i in range(firstlen):
    start = first.pop()
    end = second.pop()
    source_real[start:end+1 ] = []

print(source_real)
```

```python
#js,css 태그 제거
first = []
second = []
status = False
for i in source_real:
    
    if "<script" in i:
        if status == False:
            first.append(source_real.index("<script"))
        source_real[source_real.index("<script")] = '@'
        status = True
    if "</script>" in i:
        if status == True:
            second.append(source_real.index("</script>"))
        source_real[source_real.index("</script>")] = '%'
        status = False

    if "<style" in i:
        if status == False:
            first.append(source_real.index("<style"))
        source_real[source_real.index("<style")] = '@'
        status = True
    if "</style>" in i:
        if status == True:
            second.append(source_real.index("</style>"))
        source_real[source_real.index("</style>")] = '%'
        status = False



firstlen = len(first)

for i in range(firstlen):
    start = first.pop()
    end = second.pop()
    source_real[start:end+1 ] = []

print(source_real)
```

```python
#html 주석 제거
first = []
second = []
status = False
for i in source_real:
    if "<!--" in i:
        if status == False:
            first.append(source_real.index("<!--"))
        source_real[source_real.index("<!--")] = '@'
        status = True
    if "-->" in i:
        if status == True:
            second.append(source_real.index("-->"))
        source_real[source_real.index("-->")] = '%'
        status = False

firstlen = len(first)

for i in range(firstlen):
    start = first.pop()
    end = second.pop()
    source_real[start:end+1 ] = []

print(source_real)
```

```python
source_real = ' '.join(source_real)
source_real = source_real.replace("/*"," /* ")
source_real = source_real.replace("*/"," */ ")
source_real = source_real.replace("<"," < ")
source_real = source_real.replace(">"," > ")

source_real = source_real.split()

#주석 제거
first = []
second = []
status = False
for i in source_real:
    if "/*" in i:
        if status == False:
            first.append(source_real.index("/*"))
        source_real[source_real.index("/*")] = '@'
        status = True
    if "*/" in i:
        if status == True:
            second.append(source_real.index("*/"))
        source_real[source_real.index("*/")] = '%'
        status = False
        
    if "<" in i:
        if status == False:
            first.append(source_real.index("<"))
        source_real[source_real.index("<")] = '@'
        status = True
    if ">" in i:
        if status == True:
            second.append(source_real.index(">"))
        source_real[source_real.index(">")] = '%'
        status = False

    
firstlen = len(first)

for i in range(firstlen):
    start = first.pop()
    end = second.pop()
    source_real[start:end+1 ] = []

print(source_real)
```

```python
source_real = ' '.join(source_real)
for i in string.punctuation:
    source_real = source_real.replace(i," ")

source_real = source_real.split()

source = []
for i in source_real:
    i = i.strip()
    if i != "":
        source.append(i)
print(source)
```

```python
d = {}
for i in source:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
```

저번 과제에서는 html코드만 제거하여 순수한 텍스트들을 찾아냈다면 이번에는 js와 css코드도 제거하였다.

먼저 js와 css 코드는 <script></script>태그와 <style></style>사이에 삽입되어있다. 

따라서저 태그들과 함께 내부의 코드들도 제거해 주어야한다.

그러면 순수한 텍스트만이 남는다.

그런다음 구두문자들을 제거하기위해서 맨처음에는 일일이 인덱스를 찾아서 접근하였지만 

실수로 제거하여 더 간단한 방법을 생각해 본 결과, 그냥 join()함수를 이용하여 문자열로 만들어 준다음 replace()함수를 이용하여 

모든 구두문자를 공백으로 바꾸어주었다.

그런 다음 다시 split()함수를 이용하여 공백으로 나누어주고 빈글자가 있는 경우 제거해주었다. 

그러면 최종적인 단어들이 남게 된다.

그 후 단어의 갯수를 카운트하기 위해서 빈사전을 만들어주고, 사전에 그 키 값이 없을 경우 사전에 추가한 후 값을 1로 해주며, 

그 키가 있을 경우에는 키에 해당하는 값을 +1해주어 반복되었다는 것을 표시해준다.
