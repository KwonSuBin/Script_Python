### 문제3번

다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def setVal(self,val):
        self.count = val
    def __str__(self):
        return str('[Count(step: %s)] %s' %(self.step, self.count))
    def incr(self):
        self.setVal(self.count+self.step)
    def __call__(self):
        self.incr()
    def __add__(self,other):
        self.count = self.count+other
        return self
    def __sub__(self, other):
        self.count = self.count-other
        return self
    def __gt__(self, other):
        return self.count > other
    def __lt__(self, other):
        return self.count < other
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other
```

```python
c = Counter(10)
d = Counter(10,2)
print(c)
print(d)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c = c+5
d = d-5
print(c)
print(d)
print(c > 10)
print(d > 10) 
print(c < 10) 
print(d < 10)
print(c == 17)
print(d != 9)
```

클래스를 생성하면 많은 내장함수들이 만들어져 있다. 

그중에 대표적인 것이 __init__, __str__, __eq__등 많이 있는데 이러한 함수들을 다시 클래스에 재정의하면

원하는 대로 동작하도록 만들 수 있다.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)


if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

class내에 정의된 __init__함수는 객체를 생성할 때 실행되는 함수로 매개변수로 받은 list의 원소들을 for문을 통해서 모두 넣어준다. 

__init__함수안에서 eliminate_duplicate()라는 static 함수를 호출한다.

eliminate_duplicate() 내부에서는 for문과 not in문을 이용해서 중복되는 원소들은 하나만 추가할 수 있도록 만들어준다. 

class MySet의 객체 s를 만들면 s는 중복되지 않은 리스트의 원소를 의미한다. 

그리고 s를 print하면 내부적으로 __str__ 함수가 호출된다. 

__str__에는 출력하는 형식이 정의되어 있다.

### 문제5번

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __or__(self,other):
        li = self
        for i in other :
            if i not in li:
                li.append(i)
        return list
    def __and__(self, other):
        li = []
        for i in self:
            if i in other:
                if i not in li: 
                    li.applend(i)
        return list
                
    def __sub__(self,other):
        li = self
        for i in self:
            if i in other:
                li.pop(i)
        return li

    @staticmethod
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
```

>클래스에는 특수한 함수들이 정의되어 있는데, 이 문제에서 쓰인 함수들은 __init__,__str__,__or__,__and__,__sub__ 이다. 

>__init__함수는 클래스의 객체를 만들때 호출되는 생성자이다.

>__str__함수는 print()를 호출하면 불려지는 함수이다.

>이를 통해 출력형식을 바꿀 수 있다. 

>__or__함수는 | 기호를 사용하면 불려지는 함수이다.

>이 함수를 재정의하여 합집합을 나타낼 수 있도록 하였다. 

>__and__함수는 & 기호를 사용하면 불려지는 함수이다.

>이 함수를 재정의하여 교집합을 나타낼 수 있도록 하였다. 

>__sub__함수는 - 기호를 사용하면 불려지는 함수이다.

>이 함수를 재정의하여 차집합을 나타낼 수 있도록 하였다.
