```python
s = s.split('\n')
s.sort()
s = '\n'.join(s)
print(s)

f = open('s1.txt','w')
f.write(s)
f.close()
```

```python
f = open('s.txt','r')
s = f.read()
print(s)
```

 - 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
front = []
back= []
s = s.split()
while(s):
    front.append(s.pop(0))
    back.append(s.pop(0))

#print(front)
#print(back)
word = dict(zip(front,back))    

word_val=list(word.values())
word_val.sort()

s2 = ""
for j in range(len(word_val)):
    for i in word.keys():    
        if word[i] == word_val[j]:
            s2 += i+ ' '+ word_val[j]+'\n' 
print(s2)
```

```python
f = open('s2.txt','w')
f.write(s2)
f.close()

f = open('s.txt','r')
s = f.read()
print(s)
```

- 3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다

```python
s3 = ""
s = s.split()
cnt = 0
while(s):
    s3 += s.pop(0) +' '
    cnt += 1
    if(cnt > 2): 
        s3 += '\n'
        cnt = 0

print(s3)
```

```python
f = open('s3.txt','w')
f.write(s3)
f.close()
```

파일에 입력된 내용을 읽어오기 위해 open함수를 이용하여 파일을 읽기모드로 열고 read()함수를 이용하여 내용을 읽어온다. 


> 1) 읽어온 문자열을 sort()함수를 이용하여 정렬해준다.

> 2) 읽어온 문자열을 split()함수를 이용하여 단어로 구분해준 다음 홀수번째의 단어와 짝수번째 단어를 나누어 리스트에 넣어준다.

 > 그런 zip()함수를 이용하여 사전으로 만든 다음 values()함수를 이용하여 리스트를 만들어 정렬해 준다.

 > 정렬한 values()값들을 이용하여 keys()값들을 이어준다.

 > 여기서 사전은 key값을 이용해서 value값을 찾을 수 있지만, value값을 이용해서 key값을 찾을 수 없다.

 > 따라서 다른 방법으로 찾아주어야 하는데 그중 이중 for문을 이용하여 하나하나 비교해주었다. 

> 3) 읽어온 문자열을 split()함수를 이용하여 단어로 나누어 주었다. 

 > 그런다음 그 리스트를 pop()함수를 이용하여 처음부터 하나씩 꺼내어 문자열 더해주었다. 
 
 > 그리고 세단어를 받아온 후 문자열에 '\n'을 더해주어 다른 줄에 출력되도록 하였다.

### 문제 3번

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

```python
def frange(start=0.0, stop = -2, step = 0.1):
    if stop < 0:
        return frange(0.0,start,0.1)
    else:
        li = []
        num = start
        while stop > num:
            num = round(num,5)
            li.append(num)
            num +=step
        print(li)
frange(0.5)
frange(1.0, 2.0)
frange(2.2, 4.0, 0.5)
```

### 문제 4번

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

```python
def sum(*args):
    sum = 0
    for i in args:
        sum += i
    print(sum)
```

```python
sum(0)
sum(1,2)
sum(1,2,3,4,5)
sum(1,5,7,2,-10)
```

### 문제 5번

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

```python
def myinitial(s):
    li = list(map(lambda s : s[0].title(),s.split()))
    li = ''.join(li)
    print(li)
myinitial("as soon as possible")
```

### 문제 6번

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

```python
from functools import reduce
```

```python
def myfact(n):
    print(reduce(lambda x,y:x*y,list(range(1,n)),n))
```

```python
myfact(971)
```

```python
def myfact(n):
    if n==1:
        return 1
    if n > 1:
        return n * myfact(n-1)
print(myfact(5))
print()
print(myfact(971))
```

### 문제 7번

 import string 과 from string import * 의 차이점을 설명하시오.

> import string은 string 모듈에 있는 모든 변수나 함수 클래스 등이 모두 포함되어 string모듈에 있는 것들을 모두 이용할 수 있다.

> 따라서 string 모듈에 있는 함수를 부를때 string.함수이름() 이렇게 호출해야 한다.

> 하지만 이렇게 import string을 이용하면 string에 있는 모든 것들이 포함되므로 용량이 커지며 필요없는 것들까지 포함하게 된다. 

> 그래서 사용하고자 하는 것이 특정할 경우. 즉, 특정한 함수나 클래스들만 포함시키기 위해서 from string import*을 이용한다. 

> 이것은 string모듈에 있는 *만을 포함하도록 한다. 그러므로 *을 호출할 때 *만 써도 호출할 수 있다.

> 이렇게 함으로써 불필요한 것들을 포함하지 않아도 되므로 용량이 줄어든다.
