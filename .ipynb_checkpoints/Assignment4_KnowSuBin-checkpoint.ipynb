```python
s = s.split('\n')
s.sort()
s = '\n'.join(s)
print(s)

f = open('s1.txt','w')
f.write(s)
f.close()
```

```python
f = open('s.txt','r')
s = f.read()
print(s)
```

 - 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
front = []
back= []
s = s.split()
while(s):
    front.append(s.pop(0))
    back.append(s.pop(0))

#print(front)
#print(back)
word = dict(zip(front,back))    

word_val=list(word.values())
word_val.sort()

s2 = ""
for j in range(len(word_val)):
    for i in word.keys():    
        if word[i] == word_val[j]:
            s2 += i+ ' '+ word_val[j]+'\n' 
print(s2)
```

```python
f = open('s2.txt','w')
f.write(s2)
f.close()

f = open('s.txt','r')
s = f.read()
print(s)
```

- 3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다

```python
s3 = ""
s = s.split()
cnt = 0
while(s):
    s3 += s.pop(0) +' '
    cnt += 1
    if(cnt > 2): 
        s3 += '\n'
        cnt = 0

print(s3)
```

```python
f = open('s3.txt','w')
f.write(s3)
f.close()
```

```python
import os
import hashlib

def print_menu():
    print("""
    Welcome to Our Service
    1. Sign Up
    2. Sign In
    3. Quit""")
    
def setInformation():
    print("ID = ", end = "")
    Id = input()
    print("password = ", end = "")
    password = input()
    print("name = ", end = "")
    name = input()
    print("school = ", end = "")
    school = input()
    return Id,password,name,school

def make_file_list(i):
    s = i.replace('[',' ')
    s = s.replace(']',' ')
    s = s.replace(',',' ')
    s = s.replace(':',' ')
    s = s.split()
    return s

def myhash(password) :
    password_plain = "ripemd160"
    password_encrypted = hashlib.new(password_plain)
    password_encrypted.update(bytes(password,'ascii'))
    return password_encrypted.hexdigest()
```

### 문제 3번

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

```python
def frange(start=0.0, stop = -2, step = 0.1):
    if stop < 0:
        return frange(0.0,start,0.1)
    else:
        li = []
        num = start
        while stop > num:
            num = round(num,5)
            li.append(num)
            num +=step
        print(li)
frange(0.5)
frange(1.0, 2.0)
frange(2.2, 4.0, 0.5)
```

이 문제는 기본 인수값에 대해서 정확하게 알아야한다.

기본 인수값은 함수를 호출할 때 값이 들어오지 않으면 기본으로 설정해 놓았던 값이 자동으로 들어오게 된다.

이것을 이용할 때 가장 중요하게 여겨야 하는 것은 매개변수가 들어오는 순서이다. 

만약 여러개의 매개변수를 받아야 하는 함수에서 기본 인수값을 지정해주지않은 매개변수가 가장 먼저오게 되면 인터프리터는 이 값이 어떤 값에 들어오는지 알지 못하게 된다.

즉, 이것이 어떤 값인지 명확하지 않다.

이러한 점을 해결하기 위해서 기본 인수값이 있는 매개변수는 없는 매개변수보다 뒤쪽에 위치해야 한다.

그렇지 않을 경우 에러가 발생하게 된다.

그런데 이 문제에서는 기본인수값이 없는 변수를 앞에 위치하게 하면서 오류가 발생하지 않고 정상적으로 발생하는 것을 원했다. 

그래서 생각해낸 방법은 함수 안에서 다시 함수를 부르도록(재귀적으로) 하였다. 

stop이라는 매개변수는 기본인수값을 지정하라는 말이 없었지만, 기본 인수값이 없이는 두번째로 올수 없으므로 절대 들어올일 없는 음수값으로 지정해주었고,

stop이 음수일 경우 start값에 지정되어 있는 값을 stop에 넣어주고 다시 함수를 호출해주었다.

### 문제 4번

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

```python
def sum(*args):
    sum = 0
    for i in args:
        sum += i
    print(sum)
```

```python
sum(0)
sum(1,2)
sum(1,2,3,4,5)
sum(1,5,7,2,-10)
```

함수의 매개변수의 값이 정확하지 않을 경우에는 가변인수 리스트를 이용해주면 쉽게 할 수 있다.

sum함수는 for문을 이용해서 모든 튜플값을 더해주면 된다. 

### 문제 5번

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

```python
def myinitial(s):
    li = list(map(lambda s : s[0].title(),s.split()))
    li = ''.join(li)
    print(li)
myinitial("as soon as possible")
```

이 문제를 간단하게 해결하기 위해 map()함수와 lambda()함수를 같이 이용해 주었다. 

map()의 인수로는 (함수,시퀀스 자료형)이 들어가게 되는데, 시퀀스 자료형에는 문자열 s를 단어로 나누어진 리스트를 넣어주기 위해서

s.split()을 넣어주었고, 함수에는 각 단어의 첫글자를 대문자로 만들어주기 위해 title()함수를 써주었다. 

### 문제 6번

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

```python
from functools import reduce
```

```python
def myfact(n):
    print(reduce(lambda x,y:x*y,list(range(1,n)),n))
```

```python
myfact(971)
```

```python
def myfact(n):
    if n==1:
        return 1
    if n > 1:
        return n * myfact(n-1)
print(myfact(5))
print()
print(myfact(971))
```

이 문제를 시작했을 때 맨처음에는 reduce()함수를 이용해주면 짧고 명확하게 작성할 수 있을 것 같아서 reduce()함수를 이용하고자 했다.

### 문제 7번

 import string 과 from string import * 의 차이점을 설명하시오.

> import string은 string 모듈에 있는 모든 변수나 함수 클래스 등이 모두 포함되어 string모듈에 있는 것들을 모두 이용할 수 있다.

> 따라서 string 모듈에 있는 함수를 부를때 string.함수이름() 이렇게 호출해야 한다.

> 하지만 이렇게 import string을 이용하면 string에 있는 모든 것들이 포함되므로 용량이 커지며 필요없는 것들까지 포함하게 된다. 

> 그래서 사용하고자 하는 것이 특정할 경우. 즉, 특정한 함수나 클래스들만 포함시키기 위해서 from string import*을 이용한다. 

> 이것은 string모듈에 있는 *만을 포함하도록 한다. 그러므로 *을 호출할 때 *만 써도 호출할 수 있다.

> 이렇게 함으로써 불필요한 것들을 포함하지 않아도 되므로 용량이 줄어든다.

```python
file_list = ['www.nytimes.com.words.frequency.txt',
             'www.cnn.com.words.frequency.txt',
             'www.apple.com.frequency.txt',
             'stackoverflow.com.frequency.txt',
             'www.billboard.com.frequency.txt']

f1 = open('www.nytimes.com.words.frequency.txt','rb')
s1 = pickle.load(f1)
f1.close()

f2 = open('www.cnn.com.words.frequency.txt','rb')
s2 = pickle.load(f2)
f2.close()

f3 = open('www.apple.com.frequency.txt','rb')
s3 = pickle.load(f3)
f3.close()

f4 = open('stackoverflow.com.frequency.txt','rb')
s4 = pickle.load(f4)
f4.close()

f5 = open('www.billboard.com.frequency.txt','rb')
s5 = pickle.load(f5)
f5.close()

#print(s1['skip'])
similarity = {}


search_key = input()

for i in s1:
    if search_key == i:
        similarity['1'] = s1[search_key]
        
for i in s2:
    if search_key == i:
        similarity['2'] = s2[search_key]

for i in s3:
    if search_key == i:
        similarity['3'] = s3[search_key]
        
for i in s4:
    if search_key == i:
        similarity['4'] = s4[search_key]
        

        
for i in range(1,len(file_list)+1):
    if str(i) not in similarity:
        similarity[str(i)] = 0

print(similarity)

for i in similarity:
    print(file_list[int(i)-1])
```

본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 3가지 이상 제시하시오.

- 검색된 문장을 통해서 문맥을 파악하여 사용자가 원하는 것들이 가장 먼저오도록 한다. 
- 검색된 문장을 통해서 의도( 정보 수집을 위한 것인지, 상업적인 의도인지 등..)를 파악하여 관련된 내용들이 나오도록 한다.
- 사람들이 많이 읽은 것들(조회수가 높은 것들)이 먼저 나오도록 한다. 

먼저 영어 불용어를 제거하는 것은 불용어 리스트를 만든 후 이중 for문을 이용하여 비교해주어 같을경우 제거해주었다.

그런 다음 파일입출력을 이용하여 웹의 url과 크롤링한 단어출현빈도 사전을 저장하였다.

그 사전파일을 읽어온 후 정렬하여 단어의 개수가 많은 것 중 가장 큰 3개의 단어를 리스트로 만들어 출력하였고,

검색어를 입력하였을때 가장 많이 나온 사이트 순서대로 출력하도록 하였다. 

여기서 유사도는 매우 중요한 역할을 하는데 유사도에 따라 사이트 순서가 다르게 출력된다.

내가 정한 유사도의 기준은 단어의 개수이다.

#### 과제를 마치면서..

> 이번과제는 어려웠고 그만큼 오래걸렸다. 

> 처음보는 오류들이 많아서 구글링을 많이 하여 해결할 수 밖에 없었다. 

> 가장 난감했던 문제는 2번문제에서 암호화였다.

> 일단 처음보는 라이브러리였고, 그 안에 있는 함수들 또한 해석할 수 없었다. 

> 그래서 검색하는데 오래걸렸던 것 같다.

> 그리고 마지막 문제인 웹크롤링의 조건인 검색이 가장 어려웠다. 

> 조금 노가다를 많이 한 것 같아서 내 코드가 마음에 들지 않지만 다른 방법이 떠오르지 않았다. 

> 내 생각에서의 유사도 구현은 단어의 부분만 검색하여도 그 단어를 포함하고 있는 것들에 따라 달라지게 하고 싶었지만 도무지 생각이 나지 않아서

> 완전히 일치하는 단어의 개수가 많은 순서대로 출력하였고, 없을 경우에는 그냥 차례대로 출력하였다. 

> 더 실력이 쌓인다면 다시 도전해보고 싶다.

> 오래걸린 만큼 뿌듯함이 크다!
