```python
from urllib import request
import string
 
url = 'http://www.naver.com'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
#print(source)
print("------------------------")

#공백을 없앤다
source = source.replace("\n"," ")
source = source.replace("\t"," ")
source = source.split()
source = ' '.join(source)
print(source)
        
```
